# What is the Model Learning?

The goal of Linear Regression is to learn the relationship between the input features and the target variable:

$$
\hat{y} = w_1x_1 + w_2x_2 + w_3x_3 + w_4x_4 + w_5x_5 + b
$$

where:

- $\hat{y}$ = Predicted Disease Score
- $x_1$ = Age
- $x_2$ = BMI
- $x_3$ = Blood Pressure
- $x_4$ = Blood Sugar
- $x_5$ = Gender
- $w_1, w_2, \ldots, w_5$ = Feature weights (coefficients)
- $b$ = Bias (intercept)

Initially, the values of the weights ($w_1$–$w_5$) and the bias ($b$) are unknown. During training, the model learns the optimal values of these parameters to minimize the prediction error and produce the most accurate predictions.

## Step 4 : Hypothesis Function 

#### What is the Hypothesis Function?

The **Hypothesis Function** is the mathematical equation used by a Linear Regression model to predict the target value from the input features.

For multiple linear regression:

$$
\hat{y} = w_1x_1 + w_2x_2 + w_3x_3 + w_4x_4 + w_5x_5 + b
$$

where:

- $\hat{y}$ = Predicted value
- $x_1, x_2, \ldots, x_5$ = Input features
- $w_1, w_2, \ldots, w_5$ = Feature weights (coefficients)
- $b$ = Bias (intercept)

The hypothesis function **does not learn** the parameters. It simply uses the **current weights and bias** to generate predictions.

Initially, the predictions are inaccurate because the weights are not optimized. As the model trains, the weights and bias are updated, resulting in more accurate predictions.

### Matrix Form of the Hypothesis

Instead of computing the prediction for each sample individually,

$$
\hat{y} = w_1x_1 + w_2x_2 + \cdots + w_nx_n + b
$$

we can express the hypothesis in matrix form:

$$
\boxed{\hat{y} = Xw + b}
$$

where:

- $X$ = Feature matrix
- $w$ = Weight vector
- $b$ = Bias (intercept)
- $\hat{y}$ = Vector of predicted values

Using matrix multiplication enables the model to compute predictions for **all samples simultaneously**, making the implementation faster and more efficient.

````python
# Hypothesis Function

def hypothesis(X, w, b):
    """
    Computes the predicted values using
    the Linear Regression equation.

    Parameters
    ----------
    X : Feature Matrix
    w : Weight Vector
    b : Bias

    Returns
    -------
    Predicted values (ŷ)
    """

    y_hat = np.dot(X, w) + b #np.dot -> performs matrix multiplication 

    return y_hat
````

##  Step 5: Cost Function

## What is the Cost Function?

The **Cost Function** measures how well the Linear Regression model is performing by comparing the predicted values with the actual values.

It summarizes the overall prediction error into a **single numerical value**.

- **High Cost** → Poor predictions
- **Low Cost** → Better predictions
- **Cost = 0** → Perfect predictions

The objective of Linear Regression is to find the weights and bias that **minimize the Cost Function**.

---

## Why is the Cost Function Needed?

A computer cannot determine which model is better by simply looking at its predictions. The Cost Function provides a numerical measure of the total prediction error, allowing different models to be compared objectively.

During training, the learning algorithm continuously updates the model parameters to reduce this error and improve prediction accuracy.

````python
# We need to take an average error to get the total error
# But if we take them as is then they will cancel each other out, Therefore we square them 
# When squared -> Negative errors disappear and larger mistakes face bigger penalty 



# Cost Function
def compute_cost(X, y, w, b):
    """
    Computes the Mean Squared Error (Cost Function)
    for Multiple Linear Regression.
    """

    # Number of training examples
    m = len(y)

    # Predicted values
    y_hat = hypothesis(X, w, b)

    # Mean Squared Error Cost
    cost = (1 / (2 * m)) * np.sum((y_hat - y) ** 2)

    return cost
````

## Mean Squared Error (MSE)

The **Mean Squared Error (MSE)** is the most commonly used cost function for Linear Regression. It calculates the average squared difference between the predicted and actual values.

$$
J(w,b)=\frac{1}{2m}\sum_{i=1}^{m}(\hat{y}_i-y_i)^2
$$

where:

- $J(w,b)$ = Cost function
- $m$ = Number of training examples
- $\hat{y}_i$ = Predicted value
- $y_i$ = Actual value

The objective of Gradient Descent is to **minimize** this cost by finding the optimal values of the weights and bias.

````python
# Why divide by m ?
# If set A has 10 patients and set B has 1000 patients; set B has a larger cost just cause of the sample size; therefore we divide by m 

# Compute the initial cost

cost = compute_cost(X, y, w, b)

print("Initial Cost:", cost)
````

## Step 6 : Computing the Gradients (derivatives)

The cost function tells us how wrong the predictions are but does not tell us "how to improve" the model.

Suppose the current cost is:

$$
J(w,b) = 2785.64
$$

A high cost indicates that the model's predictions are inaccurate. However, the cost alone does not tell us:

- Which parameter should be updated?
- Should a parameter increase or decrease?
- How much should each parameter change?

To determine this, we compute the **gradient**.

The gradient is the derivative of the cost function with respect to each model parameter. It indicates the **direction** and **magnitude** of the updates required to reduce the cost.

For Linear Regression, we compute the gradients with respect to:

- The weights ($w$)
- The bias ($b$)

These gradients are then used by the **Gradient Descent** algorithm to update the model parameters and minimize the cost.

## Gradient Function

The **Gradient Function** computes how the cost changes with respect to the model parameters. These gradients are used by Gradient Descent to update the weights and bias.

### Gradient Equations

**Gradient with respect to the weights**

$$
\frac{\partial J}{\partial w}
=
\frac{1}{m}X^T(\hat{y}-y)
$$

**Gradient with respect to the bias**

$$
\frac{\partial J}{\partial b}
=
\frac{1}{m}\sum(\hat{y}-y)
$$

where:

- $X^T$ = Transpose of the feature matrix
- $\hat{y}$ = Predicted values
- $y$ = Actual values
- $m$ = Number of training examples

The computed gradients determine how the weights and bias should be updated to minimize the cost.

````python
# Gradient Function

def compute_gradients(X, y, w, b):
    """
    Computes the gradients of the Cost Function
    with respect to the weights and bias.
    """

    # Number of training examples
    m = len(y)

    # Predictions
    y_hat = hypothesis(X, w, b)

    # Error
    error = y_hat - y

    # Gradient for weights
    dw = (1 / m) * np.dot(X.T, error) # Computes gradient for all 5 weights simultaneously  
                                      # X.T -> transpose allows proper matrix multiplication 

    # Gradient for bias
    db = (1 / m) * np.sum(error) # not attached to any feature therefore we simply average the errors  

    return dw, db
````

````python
# testing 

dw, db = compute_gradients(X, y, w, b)

print("Weight Gradients:")
print(dw)

print()

print("Bias Gradient:")
print(db)
````

## Why Do We Need a Derivative for Every Weight?

The Linear Regression model is:

$$
\hat{y} = w_1x_1 + w_2x_2 + w_3x_3 + w_4x_4 + w_5x_5 + b
$$

Each weight ($w_1$ to $w_5$) and the bias ($b$) influences the prediction differently. Therefore, we compute a separate derivative for each parameter:

- $\frac{\partial J}{\partial w_1}$ (Age)
- $\frac{\partial J}{\partial w_2}$ (BMI)
- $\frac{\partial J}{\partial w_3}$ (Blood Pressure)
- $\frac{\partial J}{\partial w_4}$ (Blood Sugar)
- $\frac{\partial J}{\partial w_5}$ (Gender)
- $\frac{\partial J}{\partial b}$ (Bias)

These gradients determine how each parameter should be updated to minimize the cost function.

## Step 7 : Gradient descent 

- It is an optimization algorithm used to minimize the cost function <br>
- It finds the best values of weights and bias that produce the smallest prediction error <br>
- Instead of trying every possible combination of weights; start with a initial guess and gradually improves parameters <br>

The learning process follows these steps:

1. Make predictions using the current weights.
2. Calculate the Cost Function.
3. Compute the gradients.
4. Update the weights and bias.
5. Repeat until the cost stops decreasing.

The objective is to find the values of the parameters that minimize the Cost Function.

````python
# Gradient Descent Function

def gradient_descent(X, y, w, b, learning_rate, iterations):
    """
    Performs Gradient Descent to learn
    the optimal weights and bias.
    """

    # Store the cost after each iteration (we will plot this later )
    cost_history = []

    # Repeat the learning process
    for i in range(iterations): # Repeats the learning cycle a 1000 times 

        # Compute gradients
        dw, db = compute_gradients(X, y, w, b) # Which direction should each weight move 

        # Update the weights
        w = w - learning_rate * dw  # GRADIENT DESCENT EQUATION 

        # Update the bias
        b = b - learning_rate * db

        # Compute the new cost
        cost = compute_cost(X, y, w, b)

        # Store the cost
        cost_history.append(cost)

    return w, b, cost_history
````

````python
# TRAINING THE MODEL

# Hyperparameters

learning_rate = 0.0000001

iterations = 1000

# Train the model

w, b, cost_history = gradient_descent(
    X,
    y,
    w,
    b,
    learning_rate,
    iterations
) 
````

````python
# Checking if the cost decreases 

print("Initial Cost:", cost_history[0])

print("Final Cost:", cost_history[-1])
````

## Step 8 : Visualizing the learning process 

- During Gradient Descent, the model updates its weights repeatedly to reduce the prediction error.<br>
- After every iteration, we calculate the Cost Function.<br>
- Instead of looking at hundreds or thousands of cost values individually, we can plot them on a graph.<br>
- The graph allows us to observe how the Cost Function changes during training.<br>

A successful Gradient Descent algorithm should show:

- A high initial cost
- A steadily decreasing cost
- A final cost that converges to a minimum value

If the cost increases instead of decreasing, it usually indicates that the learning rate is too large.<br>

If the cost decreases extremely slowly, the learning rate may be too small.<br>

Therefore, plotting the Cost Function is one of the easiest ways to determine whether the learning algorithm is working correctly.<br>


````python
# Plot the Cost Function

plt.figure(figsize=(8,6)) # Creates an empty graph 

plt.plot(cost_history) # plot 

plt.xlabel("Iterations") # Labels the horizontal axis every point represents one gradient descent iteration 

plt.ylabel("Cost") # shows how wrong the model is 

plt.title("Cost Function During Gradient Descent")

plt.grid(True)

plt.show()
````

## R² Score

The **R² Score (Coefficient of Determination)** measures how well the model explains the variation in the target variable.

$$
R^2 = 1 - \frac{\sum (y-\hat{y})^2}{\sum (y-\bar{y})^2}
$$

where:

- $y$ = Actual values
- $\hat{y}$ = Predicted values
- $\bar{y}$ = Mean of the actual values

**Interpretation:**

- **$R^2 = 1$** → Perfect fit
- **$R^2 \approx 0.9$** → Excellent fit
- **$R^2 \approx 0$** → Poor fit
- **$R^2 < 0$** → Performs worse than predicting the mean

````python
# Calculate R² Score

ss_total = np.sum((y - np.mean(y)) ** 2)

ss_residual = np.sum((y - y_pred) ** 2)

r2 = 1 - (ss_residual / ss_total)

print("R² Score:", r2)
````
