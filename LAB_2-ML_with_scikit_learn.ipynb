## Exercise 1 : Compute co-variance matrix 
A covariance matrix is a square, symmetric mathematical table that summarizes the relationships (variances and covariances) between multiple variables in a dataset
### Step 1 : Create a matrix 

```python
X = np.array([
    [1,0,2],
    [0,1,1],
    [2,1,0],
    [1,1,1],
    [0,2,1]
])

print(X)
```

### Step 2 : Understand the shape 

```python
print(X.shape) # 5 samples (rows) ; 3 features (columns)
```

### Step 3 : Compute mean of each feature 

```python
mean = np.mean(X, axis=0)

print(mean)

# Feature 1 mean = (1+0+2+1+0)/5=0.8
# Feature 2 mean = (0+1+1+1+2)/5=1
# Feature 3 mean = (2+1+0+1+1)/5=1
```

### Step 4 : Center the data 
i.e. subtract the mean from each row 

```python
X_centered = X - mean

print(X_centered)
```

### Step 5 : Compute co-variance matirx using matrix multiplication 



```python
n = X.shape[0]

covariance = (X_centered.T @ X_centered)/(n-1)

print(covariance)
```

### Step 6 : Verify using Numpy 

```python
# numpy already has a covariance function 
cov_numpy = np.cov(X, rowvar=False)

print(cov_numpy)
```

### What does covariance mean ?
It tells us how 2 features vary together 
<br>
<br>
Positive Covariance : Feature A ↑ ; Feature B ↑
<br>
Negative Covariance : Feature A ↑ ; Feature B ↓
<br>
Near Zero : No relationship


## Exercise 2 : Dot Product 
Dot product is important in ML 
<br>
For example : Linear regression 
<br>
y = x*(theta) : prediction is actually a dot product 
### Step 1 : create vectors 

```python
x = np.array([2,1,2])

y = np.array([1,2,2])
```

### Step 2 : Compute dot products 

```python
dot_product = np.dot(x,y)

print(dot_product)
```

```python
# CASE 1 : Same direction 

a = np.array([1,1])
b = np.array([2,2])

print(np.dot(a,b))

# CASE 2 : Perpendicular direction 

a = np.array([1,0])
b = np.array([0,1])

print(np.dot(a,b))

# CASE 3 : Opposite direction 

a = np.array([1,1])
b = np.array([-1,-1])

print(np.dot(a,b))
```
