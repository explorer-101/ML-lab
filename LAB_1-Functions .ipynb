## Exercise 1 : Compute A(T).A

### Step 1 : create array

```python
import numpy as np

# creating an array of arrays (2D array)
A = np.array([
    [1, 2, 3],
    [4, 5, 6]
])

print(A)
```

### Step 2 : perform Transpose  

```python
AT = A.T

print(AT)
```

### Step 3 : Multiply 

To multiply two matrices, the number of columns in the first matrix must match the number of rows in the second

```python
ATA = AT @ A

print(ATA)
```

## Exercise 4 : Gaussian PDF 
### Step 1 : Create x values initialize mean and standard deviation 

```python
mean = 0
sigma = 15

x = np.linspace(-100, 100, 100)
```

### Step 2 : Compute the PDF equation 

```python
pdf = (1/np.sqrt(2*np.pi))*np.exp(-(x**2)/2) # np.exp() = means e power x 
```

### Step 3 : Plot 

```python
# Gaussian PDF formula
plt.plot(x, pdf)
plt.xlabel("x")
plt.ylabel("Gaussian PDF")
plt.title("Gaussian PDF (mean=0, sigma=15)")
plt.grid(True)
plt.show()
```

## Exercise 7 : Linear Model 

```python
# Coefficients represented as 'theta'
theta = np.array([
    [2],
    [3],
    [3]
])
```
